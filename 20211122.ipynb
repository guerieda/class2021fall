```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

C = 261.63
Db = 277.18
D = 293.66
Eb = 311.13
E = 329.63
F = 349.23
Gb = 369.99
G = 392.00
Ab = 415.30
A = 440.00
Bb = 466.16
B = 493.88
```

```python
#반짝반짝 작은별

notes = [ C, C, G, G,A,  A, G ]

durs = [ 0.5, 0.5, 0.5, 0.5, 0.5, 0.5,  1 ]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```
